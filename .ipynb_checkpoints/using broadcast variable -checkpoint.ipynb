```python
import findspark
findspark.init(0)

from pyspark import SparkConf,SparkContext

conf=SparkConf().setMaster('local').setAppName('movie-by-name')
sc=SparkContext(conf=conf)
```

```python
def movieDictionary():
    dic=dict()
    with open("/Users/Prince.Bharti/Desktop/data/spark-course/ml-100k/u.item",'r',encoding='latin-1')  as file:
        for line in file:
            li=line.split('|')
            dic[li[0]]=li[1]
    return dic
```

```python
broad_cast=sc.broadcast(movieDictionary())


base_rdd=sc.textFile("file:///Users/Prince.Bharti/Desktop/data/spark-course/ml-100k/u.data")
cleaned_rdd=base_rdd.map(lambda element:(element.split('\t')[1],(int(element.split('\t')[2]))))
reduced_result=cleaned_rdd.reduceByKey(lambda x,y:x+y)
flipped_rdd=reduced_result.map(lambda element:(element[1],element[0]))
sorted_result=flipped_rdd.sortByKey(ascending=False)
result=sorted_result.mapValues(lambda value:broad_cast.value[value]).collect()

for line in result:
    print(line)
```
