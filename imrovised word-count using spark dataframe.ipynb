```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
import re
```

```python
spark=SparkSession.builder.appName('word-count').getOrCreate()
```

```python
sc=spark.sparkContext
```

```python
base_rdd=sc.textFile("file:///Users/Prince.Bharti/Desktop/data/spark-course/Book.txt")
```

```python
def mapper(line):
    li=re.findall(r'\w+',line.lower())
    return li
    
```

```python
words_rdd=base_rdd.flatMap(mapper)
```

```python
schema_rdd=words_rdd.map(lambda element: Row(word=element))
```

```python
df=spark.createDataFrame(schema_rdd)
```

```python
result_df=df.groupBy('word').count().orderBy('count',ascending=False)
```

```python
for element in result_df.collect():
    print("the word '{}' has count {}".format(element[0],element[1]))
```
